```python
import requests
from bs4 import BeautifulSoup

# 1. 대상 URL 설정
url = "https://news.naver.com/section/105" # IT/과학 뉴스

# 2. HTTP GET 요청 (User-Agent 설정은 필수 - 봇 차단 방지)
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36"
}

response = requests.get(url, headers=headers)
```

```python
# 3. HTML 파싱

soup = BeautifulSoup(response.text, "html.parser")

# 4. 데이터 추출 (CSS Selector 활용)

# 네이버 뉴스 구조에 따라 class명은 변경될 수 있음

news_list = soup.select(".sa_text_strong")

print(f"--- 오늘의 IT 뉴스 헤드라인 ({len(news_list)}건) ---")

for idx, news in enumerate(news_list[:10], 1):

    print(f"{idx}. {news.get_text().strip()}")
```

```python
# 기사제목, 기사요약, 신문사, 작성날짜, 링크
# 하나의 태그만 가져와서 뽑고 select_one -> select

article = soup.select_one(".block1")
    

title = article.select_one('.articleSubject > a').text

summary = article.select_one('.articleSummary').contents[0].text.strip()

press = article.select_one('.press').text.strip()
wdate = article.select_one('.wdate').text
link =  article.select_one('.articleSubject > a').attrs['href']
article_id = link.split('=')[1].split('&')[0]
office_id = link.split('office_id=')[1].split('&')[0]

link = f"https://n.news.naver.com/mnews/article/{office_id}/{article_id}"

print(f'"title": {title},"summary": {summary}, "press": {press}, "wdate": {wdate}, "link": {link}' )
```

```python
import requests

url = 'https://apis.data.go.kr/1613000/RTMSDataSvcAptTrade/getRTMSDataSvcAptTrade'
params = {
      'LAWD_CD': '11110',
      'DEAL_YMD': '201801',
      'serviceKey': ''
    }

response = requests.get(url,params = params)
response
```

```python
import xmltodict

response_data = xmltodict.parse(response.content)

response_data
```

```python
items = response_data['response']['body']['items']['item']
print(items[0])

# 아파트명, 법정동, 거래금액, 전용면적, 층, 건축년도
item = items[0]
{'아파트명': item['aptNm'], '법정동': item['umdNm'], '거래금액': item['dealAmount'], '전용면적': item['excluUseAr'], '층': item['floor'], '건축년도': item['buildYear']}

```

```python
apt_data = []
for item in items:
    apt_data.append({'아파트명': item['aptNm'], '법정동': item['umdNm'], '거래금액': item['dealAmount'], '전용면적': item['excluUseAr'], '층': item['floor'], '건축년도':item['buildYear'] })
apt_df = pd.DataFrame(apt_data)
apt_df.to_csv('data/동탄아파트실거래가202501.csv')
```
