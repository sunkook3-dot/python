```python
def max_num(a, b, c):
    # 가장 큰 수를 반환하는 함수
    print(a, b, c)
    max_val = a
    if b > max_val:
        max_val = b
    if c > max_val:
        max_val = c
    return max_val 
```

```python
def min_num(num_list):
    min_val = num_list[0]
    for i in num_list:
        if min_val > i:
            min_val = i
    return min_val
```

```python
min_num([52, 273, 32, 103, 275, 24,57]) #24
min_num([-1, -273, -32, -103, -275, 24, -57]) #24
```

```python
def add_mul(choice, *args):
    # print(choice, args)
    if choice == 'add':
        result = 0
        for i in args:
            result += i
    elif choice == 'mul':
        result = 1
        for i in args:
             result *= i

    else:
        print('연산자를 똑바로 입력하세요')
        result = None
    return result
```

```python
def is_prime_num(num):
    isprime = True
    for i in range(2, num):
        if num % i == 0:
          isprime = False
          break

    return isprime
```

```python
def is_prime(num_list):

    result = []
    for num in num_list:
        isprime = True
        for i in range(2, num):
            if num % i == 0:
                isprime = False
                break

        result.append(isprime)


    return result
```
